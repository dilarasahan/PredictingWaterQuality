```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium
from folium import plugins
```

```
data = pd.read_csv("finalData.csv")
```

```
roi = data.sample(n=500,random_state=1)
```

```
roi.shape
```

```
map_hooray = folium.Map(location=[roi["lat"].iloc[0], roi["lon"].iloc[0]],
                    zoom_start = 11) # Uses lat then lon. The bigger the zoom number, the closer in you get
for i in range(roi.shape[0]):
  folium.Marker([roi["lat"].iloc[i], roi["lon"].iloc[i]], popup=f'Mean: {roi["resultMeanValue"].iloc[i]}').add_to(map_hooray)                    
map_hooray # Calls the map to display
```

```
map_hooray = folium.Map(location=[roi["lat"].iloc[0], roi["lon"].iloc[0]],
                    zoom_start = 11) # Uses lat then lon. The bigger the zoom number, the closer in you get
data = np.array(
    [
     roi["lat"].values,
     roi["lon"].values
    ]
).T
popups = roi["resultMeanValue"].values
plugins.MarkerCluster(data,popups=popups).add_to(map_hooray)                    
map_hooray # Calls the map to display
```
